```python
# -- SMALL DATA SET USED IN THAT ----

df_small = df.head(10)

plt.figure(figsize=(12,8))

# 1. Line chart (trend)
plt.subplot(2,2,1)
sns.lineplot(x=df_small["date"], y=df_small["petrol_usd_liter"])
plt.title("Petrol Trend")

# 2. Bar chart (country comparison)
plt.subplot(2,2,2)
sns.barplot(x=df_small["country"], y=df_small["petrol_usd_liter"])
plt.title("Country vs Petrol")

# 3. Histogram (distribution)
plt.subplot(2,2,3)
sns.histplot(df_small["petrol_usd_liter"], bins=5)
plt.title("Petrol Distribution")

# 4. Scatter plot (relationship)
plt.subplot(2,2,4)
sns.scatterplot(x=df_small["brent_crude_usd"], y=df_small["petrol_usd_liter"])
plt.title("Crude vs Petrol")

plt.tight_layout()
plt.show()
```

# The dataset analysis reveals strong relationships among fuel prices and influencing factors. Petrol, diesel, and LPG prices show a very high positive correlation, indicating they move together and are likely influenced by similar underlying factors.

# Brent crude oil prices exhibit a moderate positive correlation with fuel prices, suggesting that global oil prices impact domestic fuel costs, but not entirely due to external factors such as taxation and policies.



# Tax percentage shows a noticeable impact on fuel prices, highlighting its role as a key contributor to price variation. Time-based analysis indicates gradual changes in fuel prices over periods, reflecting market trends and policy adjustments.

# Categorical analysis using country data did not show significant linear correlation, but it helps in comparative analysis across regions.

    

# Overall, the dataset demonstrates that fuel prices are primarily driven by crude oil trends and taxation, with strong interdependence among different fuel types.
